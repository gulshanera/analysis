```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RANSACRegressor
from sklearn.metrics import r2_score
import pandas as pd
import os
import numpy as np

# Load the data into a dataframe
# df = pd.read_csv('data.csv')
for i in range(4,13):
    # Extract the predictor and response variables
#     x = ndf[f"S{i}"].values.reshape(-1, 1)
    x=ndf.iloc[:,i].values.reshape(-1,1)
    y=ndf['BAMPM2.5'].values

    # Fit a linear regression model and a robust regression model to the data
    model_1 = LinearRegression().fit(x, y)
    y1_pred = model_1.predict(x)
    r2 = r2_score(y, model_1.predict(x))
    mod = RANSACRegressor().fit(x, y)
    y2_pred = mod.predict(x)

    # Calculate the predicted values for each model
   
   
    
   
    rr2= r2_score(y, mod.predict(x))
    # Create a figure and a subplot
    fig, ax = plt.subplots()

    # Plot the data and the regression lines
    ax.scatter(x, y, label='Data')
    ax.plot(x, y1_pred, label='Linear regression')
    ax.plot(x, y2_pred, label='Robust regression')

    # Add a legend and labels
    ax.legend(loc='lower right')
    ax.set_xlabel(f"{ndf.columns[i]}")
    ax.set_ylabel('BAMPM2.5')
    # Add a label showing the R^2 value
    ax.text(0.05, 0.90, f'R^2 = {r2:.2f}', transform=ax.transAxes)
    ax.text(0.05, 0.85, f'RR^2 = {rr2:.2f}', transform=ax.transAxes)
    plt.savefig('ndf.columns[i].jpg',dpi=400)
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Read data from a csv
df = pd.read_excel("D:/air quality/OneDrive - design.iitd.ac.in/Desktop/New folder/Jodhpur data/one_outer.xlsx")
df = df.dropna()

# Extract data from columns
x = df['AT']
y = df['AH']
z = df['S1']

# Create a meshgrid
x_unique = np.unique(x)
y_unique = np.unique(y)
X, Y = np.meshgrid(x_unique, y_unique)

# Reshape the z values to match the meshgrid shape
if len(x_unique) * len(y_unique) == len(z):
    Z = z.reshape((len(y_unique), len(x_unique)))
else:
    print('Error: the shape of z does not match the shape of the meshgrid')
    Z = np.zeros((len(y_unique), len(x_unique)))

# Create the surface plot
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])

# Set the axis labels
fig.update_layout(scene=dict(xaxis_title='Reference Sensor', yaxis_title='Humidity', zaxis_title='Temperature'))

# Show the figure
fig.show()
```
