```python

import pandas as pd
from scipy import stats

# Assuming the dataset is stored in a DataFrame called 'data' and the column name is 'column_name'
data = pd.read_excel("C:/Users/gulsh/Downloads/df.xlsx")
column_names = ['P_D2020', 'income level', 'landcover_percentage', 'mean_temp', 'Residential_area_percentage']

for column in column_names:
    data[column + '_zscore'] = stats.zscore(data[column])

data.to_excel("C:/Users/gulsh/Downloads/df_zscore.xlsx", index= False)
```

```python
import pandas as pd
import numpy as np
data = pd.read_excel("C:/Users/gulsh/Downloads/df_zscore.xlsx")
# Assuming the z-scores are stored in columns 'column1_zscore', 'column2_zscore', etc.
zscore_columns = ['P_D2020_zscore', 'income level_zscore', 'landcover_percentage_zscore', 'mean_temp_zscore', 'Residential_area_percentage_zscore']

# Calculate HVI component scores for each z-score column
for zscore_column in zscore_columns:
    data[zscore_column.replace('_zscore', '_hvi')] = np.where(data[zscore_column] <= -2, 1,
                                                             np.where(data[zscore_column] <= -1, 2,
                                                                      np.where(data[zscore_column] <= 0, 3,
                                                                               np.where(data[zscore_column] <= 1, 4,
                                                                                        np.where(data[zscore_column] <= 2, 5, 6)))))

# Print the updated DataFrame
data.to_excel("C:/Users/gulsh/Downloads/df_zscore_hvi.xlsx", index=False)
```
